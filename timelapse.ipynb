```python
# process json
import json

bldg = 'clark'

with open('C:/Users/aspen/timelapse/'+bldg+'/datetimessupreme.json') as f:
    data = json.load(f)
    
new_dts = {}
for date, times in data.items():
#     print(date)
    timelist = []
    for hour, mms in times.items():
#         print(hour)
        test1 = list(mms.items())
#         print(test1)
        test1.sort(key=lambda x: int(x[0]))
#         print(test1)
        for minute, seconds in test1:
#             print(minute)
#             print(seconds[0])
            timelist.append(hour.zfill(2) + minute.zfill(2) + str(seconds[0]).zfill(2))
#             print(timelist)

#         print(mms)
#     print(times)
    new_dts[date] = timelist

len(new_dts)
```

```python
# generate video with all images
import imageio
import os

filenames = os.listdir('images')

print(len(filenames))

with imageio.get_writer('timelapse4-60fps.mp4', mode='I', fps=60) as writer:
    for i, filename in enumerate(filenames[30000:40000]):
        print(str(i) + ' - ' + filename, end='\r')
        try:
            image = imageio.imread('images/' + filename)
            writer.append_data(image)
        except Exception as e:
            print(str(i) + ' - ' + filename)
            print(e)
```
